```python
# 初期化
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams["font.family"] = "serif"
plt.rcParams["font.size"] = 16
plt.rcParams["text.usetex"] = False
```

```python
# データ読み込み
df = pd.read_csv("results/analyzed_models.csv")
df["ratio"] = df["zeta"] / df["eta"]
```

```python
# r と zeta/eta の関係
plt.scatter(x=df["ratio"], y=df["r"])
plt.xlabel("ζ/η")
plt.ylabel("r")
```

```python
# g と zeta/eta の関係
plt.scatter(x=df["ratio"], y=df["g"])
plt.xlabel("ζ/η")
plt.ylabel("g")
```

```python
# y と zeta/eta の関係
plt.scatter(x=df["ratio"], y=df["y"])
plt.xlabel("ζ/η")
plt.ylabel("y")
```

```python
# h と zeta/eta の関係
plt.scatter(x=df["ratio"], y=df["h"])
plt.xlabel("ζ/η")
plt.ylabel("h")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(100)
y1 = x * x
y2 = x
```

```python
plt.plot(x, y1)
plt.xlabel("This is X label")
plt.ylabel("This is Y label")
plt.savefig("test1.png")
plt.show()
plt.close()

plt.plot(x, y2)
plt.savefig("test2.png")
plt.show()
plt.close()
```
